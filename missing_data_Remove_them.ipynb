```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\pshah\\Desktop\\ml_ krishniyak\\data_set\\data_science_job.csv')
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
missing_val=[val for val in df.columns if df[val].isnull().mean() < 0.05 and df[val].isnull().mean()>0]
missing_val
```

```python
new_df=df[missing_val].dropna()
```

```python
new_df.shape
```

### now we will check if there is any change in the data distribution after removing missing val

```python
# previous data distribution and current data
plt.figure(figsize=(14,2))
plt.subplot(121)
sns.histplot(df['city_development_index'],kde=True)
sns.histplot(new_df['city_development_index'],kde=True)
plt.title('city_development_index')
plt.show()
```

```python
# above work with help of loop
for val in missing_val:
    plt.figure(figsize=(14,2))
    plt.subplot(121)
    sns.histplot(df[val],kde=True)
    sns.histplot(new_df[val],kde=True)
    plt.title(val)
    plt.show()
    
```

### from the above we can make the conclusion that there is not effect on the distribution of the data data after removing the missing value.
